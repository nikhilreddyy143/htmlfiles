```python
import pandas as pd
```

```python
df=pd.read_csv("/home/cse-lab/Desktop/iris.csv")
```

```python
df.head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
variety=df['variety'].value_counts()
variety=dict(variety)
print(variety)
k=variety.keys()
print(k)
v=variety.values()
print(v)
```

```python
fig=plt.figure(figsize=(8,8))
plt.pie(x=v,labels=k,explode=[0.1,0.1,0.1],autopct='%0.1f%%')
plt.title("Iris species%")
plt.legend()
plt.show()
```

```python

x7=df['sepal.width']
y7=df['sepal.length']
fig=plt.figure(figsize=(5,5))
plt.scatter(y7,x7)
plt.xlabel('sepal.width',color="blue")
plt.ylabel('sepal.length',color="black")
plt.title("the iris Data set")
plt.show()
```

```python
sepallength=df['sepal.length']
plt.hist(sepallength)
plt.xlabel("sepal length")
plt.ylabel("frequency")
plt.show()
```
